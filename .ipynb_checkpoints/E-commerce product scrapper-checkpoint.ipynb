```python
from bs4 import BeautifulSoup
import requests
import os
import pandas as pd

url = "https://www.macmerise.com/shop/iphone-15/macmerise-glass-case"
save_directory = "image_folder" 

if not os.path.exists(save_directory):
    os.makedirs(save_directory)

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

listing_items = soup.find_all('div', class_='listing-item')

data = []
for item in listing_items:
    img_tag = item.find('img')
    if img_tag:
        image_url = img_tag.get('src')

        listing_info = item.find('div', class_='listing-info')
        if listing_info:
            name_tag = listing_info.find('h3')
            if name_tag:
                name = name_tag.text.strip()

                
                image_name = f"{name.replace(' ', '_')}.jpg"  
                image_path = os.path.join(save_directory, image_name)
                try:
                    img_response = requests.get(image_url)
                    if img_response.status_code == 200:
                        with open(image_path, 'wb') as img_file:
                            img_file.write(img_response.content)
                        data.append({'Name': name, 'Image Path': image_path})
                except Exception as e:
                    print(f"Error downloading image {image_name}: {e}")

df_images = pd.DataFrame(data)

```

```python
import matplotlib.pyplot as plt
import os
from PIL import Image

def display_images(image_paths):
    num_images = len(image_paths)
    num_cols = 5  
    num_rows = (num_images - 1) // num_cols + 1

    fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 18))

    for i, ax in enumerate(axes.flat):
        if i < num_images:
            try:
                img = Image.open(image_paths[i])
                ax.imshow(img)
                ax.axis('off')
            except OSError as e:
#                 print(f"Error loading Image {i+1}: {e}")
                ax.axis('off')  # Hide empty subplots
        else:
            ax.axis('off')  

    plt.tight_layout()
    plt.show()

image_paths = df_images['Image Path'].tolist()


display_images(image_paths)
```
